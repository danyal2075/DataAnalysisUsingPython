# DataFrame
    - Creating Dic
    - DataFrame are two dimentional data
    - df['col_name'] : it will give Series
    - df[['col_name']] : it will give two dimentional
    - df[['Name', 'course']] : to access two col

```python
# Creating Dictionary
students = {'Roll Numbers': [1,2,3,4,5], 'Name':['sohail', 'danyal', 'hamza', 'mehran', 'zakriya'], 
            'course': ['AI', 'IOT', 'BC', 'MT', 'ACCA']}
students
```

```python
# To convert Dictionary into DataFrame
df = pd.DataFrame(students)
print(f'Type of data in Dic : {type(df)}')
df
```

```python
# Creating Series. if we take just one col then it converts into Series
df['Name']
```

```python
df[['Name', 'course']]
```

```python
type(df['Name'])
```

```python
# we can also make a single col in  dataframe 
type(df[['Name']])
```

```python
type(df[['Name', 'course']])
```

```python
# Saving this file into excel data
df.to_csv("student_data.csv", index =False )
```

# Adding New Col or Data
    - Assiging single value to the all the row
    - adding via list
    - Conditional Based
        - Comprehesive method
        - Apply(lambda and np.where)
        - Via making seperate function
    - Mapping Dictionary

```python
# single value into all the rows
df['Country'] = 'Pakistan'
```

```python
df
```

```python
df.to_csv('student_data.csv', index = False)
```

```python
# Adding list 
df['Ages'] = [30,18,34,54,23]
df['Mode'] = ['Onsite', 'Online' , 'Onsite', 'Online', 'Onsite']
```

```python
df
```

```python
# Adding Conditional based Col : comprehensive
df['Status'] = ['Pass' if age > 25 else 'Fail' for age in df['Ages']]
```

```python
df
```

```python
# Conditional based adding can be done via lambda and making function
df['City'] =df['Mode'].apply(lambda x: 'Karachi' if x == 'Onsite' else 'Outside Kar') 
df
```

```python
# lets try via np.where
import numpy as np
df['Lecture'] = df['Mode'].apply(lambda x:np.where(x == 'Onsite' , 'Monday', 'Sunday'))
df
```

```python
# lets try via craeting function
def exam(x):
    if x == 'Onsite':
        return 'Hard'
    else:
        return 'Medium'   
    
# Lets apply this function
df['Exam'] = df['Mode'].apply(exam)
df
```

```python
# Mapping dic
dict = {'sohail' : 28, 'danyal' : 30 , 'zakriya' : 9, 'hamza': 22, 'mehran' : 24}
df['NewAges'] = df['Name'].map(dict)
df
```

# Adding New Row
    - Append
    - loc

```python
# Adding new row via Append
newstudent = pd.Series({
    'Roll Numbers' : 6,
    'Name' : 'Muhammad' ,
    'course':'School',
    'Country' : 'Pakistan',
    'Ages' : 14,
    'Mode' : 'Onsite',
    'Status' : 'Pass',
    'City' : 'karachi',
    'Lecture': 'Monday',
    'Exam': 'Hard',
    'NewAges' : '13'
})
df = df.append(newstudent, ignore_index=True)
df
```

```python
# Adding new row via loc
newstudent2 = pd.Series({
    'Roll Numbers' : 7,
    'Name' : 'Zaman' ,
    'course':'MS',
    'Country' : 'Pakistan',
    'Ages' : 60,
    'Mode' : 'Online',
    'Status' : 'Pass',
    'City' : 'Outside kar',
    'Lecture': 'Sunday',
    'Exam': 'Medium',
    'NewAges' : '59'
})
df.loc[8] = newstudent2
df
```
