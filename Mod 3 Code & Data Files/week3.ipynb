# Aggregation Function:
it it used to summarize the data other than avg.

we can pass the values. and then in the agg() pass the operation

we can also pass a dic, column should be the key and operation would be the value.


```python
pivot_table_agg_func = pd.pivot_table(df, index=['city', 'state'], values = ['review_count'], aggfunc = [np.sum] )
pivot_table_agg_func
```

```python
# for more details. we can specify to summarize the data on the basis of col.
pivot_table_agg_func_col = pd.pivot_table(df, index=['city', 'state'], values = ['review_count'], aggfunc = [np.sum], 
                                      columns =['take_out'] )
pivot_table_agg_func_col
```

```python
# now we can also add more than two col to the aggfunc and can perform different operations
pivot_table_aggfunc_adv = pd.pivot_table(df, index = ['city', 'state'], 
                                         aggfunc = {'review_count': np.sum, 'stars':np.mean})
pivot_table_aggfunc_adv
```

```python
pivot_table_aggfunc_adv_col = pd.pivot_table(df, index = ['city', 'state'], 
                                         aggfunc = {'review_count': np.sum, 'stars':np.mean}, columns = ['take_out'])
pivot_table_aggfunc_adv_col
```

# Histogram:
this is used for the distrubition of the values in the data.

How do review patterns differ across different cities? Which city has a higher number of five-star reviews or one-star reviews, Pittsburgh or Las Vegas? Let's find out by plotting a histogram that compares the distribution of rating scores between businesses in Pittsburgh and Las Vegas.

```python
df_pitts = df[df['city'] == 'Pittsburgh']

df_pitts
```

```python
df_vegas = df[df['city'] == 'Las Vegas']
df_vegas
```

```python
df_pitts_star = df_pitts['stars']
```

```python
df_vegas_star =df_vegas['stars']
type(df_vegas_star)
```

```python
import matplotlib.pyplot as plt
plt.hist(df_pitts_star, 
         bins = 'auto',
        label = 'Pittsburg',
        color = 'red'
        )
plt.hist(df_vegas_star, 
         bins = 'auto',
        label = 'Las Vegas',
        color = 'orange'
        )
# plt.title ('Rating of Pittsburg and Las Vegas')
plt.legend(loc = 'best')
plt.xlabel('Rating')
plt.ylabel('Number of Rating')
plt.show()
```

```python
# ploting side by side the two data
plt.hist(
    [df_pitts_star, df_vegas_star],
    bins = 'auto',
    color = ['red', 'blue'],
    label = ['Pittsburg', 'Las Vegas']

)
plt.legend( loc= 'best')
plt.xlabel('Rating')
plt.ylabel('Amoun of Rating')
# plt.title('Rating Summarization')
plt.show()
```

# Scatter plot

```python
plt.scatter(
    df_Health_medical['stars'], df_Health_medical['review_count'],
    marker = 'o',
    color = 'r',
    s = 100,
    label = 'Health and Medical'
)
plt.scatter(
    df_fast['stars'], df_fast['review_count'],
    marker = '>',
    color = 'g',
    s = 100,
    label = 'fast food'
)
plt.scatter(
    df_breakfast['stars'], df_breakfast['review_count'],
    marker = '*',
    color = 'b',
    s = 100,
    label = 'BreakFast'
)
plt.legend(loc = 'best')
plt.xlabel('Rating')
plt.ylabel('Rating Counts')
axes = plt.gca()
axes.set_yscale('log')
# plt.title('Rating of Pittsburg and Las Vegas')
plt.show()
```
