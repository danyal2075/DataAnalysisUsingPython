# Task:
we are going to count the review for 'Food' in Pittsburg City

# .str.contains()
cat0_food = sheets['category_0'].str.contains('Food')

it gives all the rows where food will be found with some other string like (Fast Food)

cat0_food = sheets['category_0'] == 'Food'

second filteration method will just filter rows containing only 'food' string. it will also drop 'Fast Food'

```python
# Creating filters
cat0_food = sheets['category_0'].str.contains('Food') 
cat1_food = sheets['category_1'].str.contains('Food')
pittsburg_food = sheets['city'] == 'Pittsburgh'
```

```python
# Applying filters
final_food = sheets[ ((cat0_food | cat1_food) & pittsburg_food)]
final_food
```

```python
# Counting Reviews
count_review_pittsburg_food = final_food['review_count'].sum()
count_review_pittsburg_food
```

# Task:
calculate the average/ mean of the auto repair shops in pittsburg using mean() method

```python
# creating filters
cat0_autoshop = sheets['category_0'].str.contains('Auto Repair')
cat1_autoshop = sheets['category_1'].str.contains('Auto Repair')
pittsburg = sheets['city'] == 'Pittsburgh'
# applying filters
final_data_auto = sheets[((cat0_autoshop |cat1_autoshop ) & pittsburg)  ]
final_data_auto
```

```python
# getting mean of the data
mean_auto = final_data_auto['review_count'].mean()
mean_auto
```
